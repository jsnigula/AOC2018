```python
import re
import numpy as np
```

```python
def find(x, y, vx, vy):

    dx = np.max(x) - np.min(x)
    dy = np.max(y) - np.min(y)

    n = 0
    while True:
    # for i in range(10):

        nx = x + vx
        ny = y + vy

        ndx = np.max(nx) - np.min(nx)
        ndy = np.max(ny) - np.min(ny)

        if ndx > dx or ndy > dy:
            break

        y = ny
        x = nx
        dy = ndy
        dx = ndx
        n += 1

    plot(x, y)

    return n

def plot(x, y):

    ny = max(y)-min(y)+1
    nx = max(x)-min(x)+1
    data = np.zeros((ny, nx), dtype=int)
    for py, px in zip(y, x):
        data[py-min(y), px-min(x)] = 1
    for py in range(ny):
        line = ''
        for px in range(nx):
            if data[py, px] == 1:
                line += '#'
            else:
                line += '.'
        print(line)
        
def read_input(infile):
    x = []
    y = []
    vx = []
    vy = []
    p = r'position=<\s*(-?\d*),\s*(-?\d*)> velocity=<\s*(-?\d*),\s*(-?\d*)>'
    with open(infile, 'r') as inf:
        for line in inf.readlines():
            d = re.findall(p, line.strip())[0]
            x.append(int(d[0]))
            y.append(int(d[1]))
            vx.append(int(d[2]))
            vy.append(int(d[3]))
    return np.array(x, dtype=int), np.array(y, dtype=int), np.array(vx, dtype=int), np.array(vy, dtype=int)
```

```python
print('*******\nPuzzle1\n*******\n')

print('\n*******\nPuzzle2\n*******\n')

print('Test case\n---------\n')

res = find(*read_input('input10a.txt'))

print(f'\nIt took {res} seconds')

assert res == 3

print('\nPuzzle case\n-----------\n')

res = find(*read_input('input10.txt'))

print(f'\nIt took {res} seconds')

assert res == 10880
```
