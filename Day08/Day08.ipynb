```python
def parse_node(nodeid, data, nodes):

    node = [[], []]
    n_childs = data[0]
    n_meta = data[1]

    data = data[2:]
    
    for c in range(n_childs):
        data = parse_node(f'{nodeid}-{c+1}', data, nodes)
        node[0].append(f'{nodeid}-{c+1}')

    for m in range(n_meta):
        node[1].append(data[m])

    nodes[nodeid] = node
    data = data[n_meta:]

    return data

def read_input(infile):
    nodes = {}
    with open(infile, 'r') as inf:
        data = [int(v) for v in inf.readline().strip().split()]

    nid = 1
    while len(data) != 0:
        data = parse_node(f'{nid}', data, nodes)

    return nodes

def sum_metadata(nodes):

    meta = 0
    for n in nodes.values():
        meta += sum(n[1])

    return meta

def value_of_node(nodeid, nodes):

    if len(nodes[nodeid][0]) == 0:
        return sum(nodes[nodeid][1])

    v = 0
    for idx in nodes[nodeid][1]:
        try:
            cnode = nodes[nodeid][0][idx-1]
            v += value_of_node(cnode, nodes)
        except (KeyError, IndexError):
            pass
    return v
    
```
